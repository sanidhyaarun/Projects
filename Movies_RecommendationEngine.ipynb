```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv(r"C:\Users\f50p8ls\Canvas_iNeuron\movie_metadata.csv")
data.head()
```

```python
data.isnull().sum(axis=0)
```

```python
data['actor_1_name']=data['actor_1_name'].replace(np.nan,"unknown")
data['actor_2_name']=data['actor_2_name'].replace(np.nan,"unknown")
data['actor_3_name']=data['actor_3_name'].replace(np.nan,"unknown")
data['director_name']=data['director_name'].replace(np.nan,"unknown")
```

```python
data.isnull().sum(axis=0)
```

```python
data['genres'].head()
```

```python
data['genres']=data['genres'].str.replace('|'," ")
```

```python
data['genres'].head()
```

```python
data['movie_title'].head()
```

```python
data['movie_title']=data['movie_title'].str.lower()
```

```python
data['movie_title'].head()
```

```python
# to remove the extra special character at the end of the movie names
data['movie_title']=data['movie_title'].str[:-1]
```

```python
data['comb']=data['actor_1_name']+" "+data['actor_2_name']+" "+data['actor_3_name']+" "+data['genres']+" "+data['director_name']
```

```python
data['comb'].head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Creating a count matrix row
cv=CountVectorizer()
count_matrix=cv.fit_transform(data['comb'])

# now similarity score matrix
sim=cosine_similarity(count_matrix)
```

### Making a function for the same

```python
def recommend(m):
    m=m.lower()
    if m not in data['movie_title'].unique():
        print("This movie is not in our database")
    else:
        i=data.loc[data["movie_title"] == m].index[0]
        lst=list(enumerate(sim[i]))
        lst=sorted(lst,key=lambda x:x[1],reverse=True)
        lst=lst[1:11]
        l=[]
        len(lst)
        # to find out index number
        for i in range(len(lst)):
            a=lst[i][0]
            l.append(data['movie_title'][a])
        # to find the name using index number
        for i in range(len(l)):
            print(l[i])
            
        
```

```python
recommend("Avatar")
```
